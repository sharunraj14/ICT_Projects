**Case Study 1, Question No. 1**

*Bank Account Manager*

```python
account_balance = 500.00
```

```python
def display_balance():
  print("\nCheck the balance")
  print(f"\nCurrent balance: {account_balance:.2f}")
```

```python
def deposit(amount):
  global account_balance
  if amount > 0:
    account_balance += amount
    print(f"\n{amount:.2f} deposited Succesfully")
    display_balance()

  else:
        print("\n Deposit Failed ")
        print("Deposit amount cannot be zero")
```

```python
def withdraw(amount):
  global account_balance
  if amount > 0:
    if account_balance >= amount:
      account_balance -= amount
      print(f"{amount:.2f} withdrawn successfully")
      display_balance()
    else:
      print("Insufficient balance")
      display_balance()
  else:
    print("Withdrawal amount cannot be zero")
```

**Outputs**

```python
display_balance()
```

```python
deposit(150.50)
```

```python
withdraw(200.00)
```

```python
withdraw(1000.00)
```

```python
deposit(0)
```

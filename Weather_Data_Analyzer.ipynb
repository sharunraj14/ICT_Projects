*Weather Data Analyzer*

```python
region = "India"
temperatures = [28.5, 30.2, 22.1, 19.5, 25.0]
humidity = [65, 45, 88, 75, 70]
```

```python
def analyze_weather_data(temp, humid):
  print(f"\n Analyzing Temp: {temp}°C, Humidity: {humid}%")

  if temp >= 30.0 and humid < 50:
    condition = "Hot and Dry"
  elif temp < 20.0 and humid > 70:
    condition = "Cool and Damp"
  elif humid > 85:
    condition = "High chance of Rain"
  else:
    condition = "Mild and Moderate"

  print(f"Condition: {condition}")
  return condition
```

```python
def average_calculation(temp, humid):

    print(f"\n  {region} ")

    if temp:
        avg_temp = sum(temp) / len(temp)
        print(f"Average Temperature: {avg_temp:.2f}°C")

    if humid:
        avg_humid = sum(humid) / len(humid)
        print(f"Average Humidity: {avg_humid:.2f}%")

    return avg_temp, avg_humid
```

**Outputs**

```python
average_calculation(temperatures, humidity)
```
